
Data Slicing

```python
data_train =pd.DataFrame(data.Close[0:int(len(data)*0.80)])
data_test =pd.DataFrame(data.Close[int(len(data)*0.80):len(data)])
```

```python
data_train.shape[0]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

MinMax scaler is used for scaling data from range (0,1) Neural networks (like LSTM) perform better when input features are on a similar scale. Large-scale differences can:

Slow down learning

Bias the model toward larger-valued features

```python
data_train_scale = scaler.fit_transform(data_train)
```

```python
x=[]
y=[]
for i in range(100,data_train_scale.shape[0]):
    x.append(data_train_scale[i-100:i])
    y.append(data_train_scale[i,0])
```

THIS COMMAND WILL TAKE THE APPROPRIATE RANGE FOR SCALING

```python
x=[]
y=[]

for i in range(100,data_train_scale.shape[0]):
    x.append(data_train_scale[i-100:i])
    y.append(data_train_scale[i,0])
    
   
```

here we are creating a data set of 100 days windows to predict next day's output like
day 0 to day 99	 =  day 100
day 1 to day 100 =	day 101
day 2 to day 101 =	day 102


now we will create a model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
```

keras is producing layers which helps in building of structure of neural networks. These are building blocks of neural networks. Output of one layer is considered as input of next layer

 converting this data set to arrays(numpy) for the input in our model
  

```python
x = np.array(x)
y = np.array(y)

# Reshape to 3D for LSTM
x = np.reshape(x, (x.shape[0], x.shape[1], 1))
```

```python
model = Sequential()  #layers are added one after another in sequential manner
model.add(LSTM(units=50, activation='relu', return_sequences=True,input_shape=(x.shape[1],1)))
model.add(Dropout(0.2))  
model.add(LSTM(units=60, activation='relu', return_sequences=True))
model.add(Dropout(0.3)) #input shape(shape of layer) is only defined in starting layer and other will have same size
model.add(LSTM(units=80, activation='relu', return_sequences=True))
model.add(Dropout(0.4))
model.add(LSTM(units=120, activation='relu',)) # here not using return seq as it is last layer won't be input for other layer
model.add(Dropout(0.5))

model.add(Dense(units=1))
```

we are using Long Short-Term Memory Model 
units=50: 50 memory units (neurons) in the LSTM layer.

activation='relu': ReLU activation function for non-linearity 
If the input x is positive, ReLU returns x.
If the input x is negative or zero, ReLU returns 0.

return_sequences=True: Tells the model to return the full sequence (used when stacking LSTMs) i.e. Output of one layer is considered as input of next layer

input_shape=(x.shape[1], 1): Input shape expected by the first LSTM

Dropout randomly disables (0.2)20% of neurons during training to prevent overfitting.
Dropout randomly breaks up the network temporarily, forcing it to learn multiple paths to the correct output. Itâ€™s like training multiple models at once, then averaging their knowledge. not dependent on 1 path or same neurons

```python
model.compile(optimizer='adam', loss='mean_squared_error')
#configuring model for training
# optimizer Adam (short for Adaptive Moment Estimation) used to update neural network weights efficiently during training.
# loss mean_squared_error=It measures the average squared difference between predicted values and true values. 

```

```python
model.fit(x, y, epochs=50, batch_size=32,verbose=1)
```

this function helps in training of the model x=input and y=target values epochs=50 refers to complete passes through the entire training dataset.
The model will go through all the x, y pairs 50 times, refining weights each time. batch size=32 ishe number of samples processed before the model updates its weights.

 past data as context for predictions, ensuring continuity in time series forecasting

```python
data_test_scale=scaler.fit_transform(data_test)
#this is used for scaling data. Stock prices can be very different from each other (some might be in hundreds, others in thousands). Without scaling, big numbers could dominate the learning process
#(0 to 1)
```

```python
x=[]
y=[]

for i in range(100,data_test_scale.shape[0]):
    x.append(data_test_scale[i-100:i])
    y.append(data_test_scale[i,0])
    
x,y = np.array(x),np.array(y)  
```
